## **1. K-means clustering**
---

#### **1.1 No Data Preprocessing**
---







```python
#for c=3
from pycaret.clustering import *
print("For Cluster = 3")
kMeanClusteringParameters3 = setup(df, verbose=False)
KMeanClusteringModel3 = create_model('kmeans', num_clusters=3)
kMeanPrediction3 = assign_model(KMeanClusteringModel3)
kMeanPrediction3

#for c=4
print("For Cluster = 4")
kMeanClusteringParameters4 = setup(df, verbose=False)
KMeanClusteringModel4 = create_model('kmeans', num_clusters=4)
kMeanPrediction4 = assign_model(KMeanClusteringModel4)
kMeanPrediction4

#for c=5
print("For Cluster = 5")
kMeanClusteringParameters5 = setup(df, verbose=False)
KMeanClusteringModel5 = create_model('kmeans', num_clusters=5)
kMeanPrediction5 = assign_model(KMeanClusteringModel5)
kMeanPrediction5
```

##### **Visulazation using Graphs,plots etc**
---





###### (a) **2D plot**


```python
plot_model(KMeanClusteringModel3, plot='cluster')
plot_model(KMeanClusteringModel4, plot='cluster')
plot_model(KMeanClusteringModel5, plot='cluster')
```

###### (b) **3D plot**

```python
plot_model(KMeanClusteringModel3, plot = 'tsne')
plot_model(KMeanClusteringModel4, plot = 'tsne')
plot_model(KMeanClusteringModel5, plot = 'tsne')
```

###### (c) **Elbow plot**

```python
plot_model(KMeanClusteringModel3, plot = 'elbow')
```

###### (d) **Silhouette Plot**


```python
plot_model(KMeanClusteringModel3, plot = 'silhouette')
plot_model(KMeanClusteringModel4, plot = 'silhouette')
plot_model(KMeanClusteringModel5, plot = 'silhouette')
```

###### (e) **Distribution Plot**

```python
plot_model(KMeanClusteringModel3, plot = 'distribution')
plot_model(KMeanClusteringModel4, plot = 'distribution')
plot_model(KMeanClusteringModel5, plot = 'distribution')
```
